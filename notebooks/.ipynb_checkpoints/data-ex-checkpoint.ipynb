```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/Home_Guards_Actual_Strength_for_the_Month_January_2024.csv")

# print("Missing Values:\n", df.isnull().sum())
 
# Fill or drop missing values
df = df.dropna()  # You can use df.fillna(value) for specific handling
 
# Check for duplicates and remove them
df = df.drop_duplicates()
 
# Display summary
# df.info()
import matplotlib.pyplot as plt
 
# Summary statistics

print(df.describe())
 
# Example: Bar chart for a categorical column

plt.figure(figsize=(10,5))

df['ACTUAL STRENGTH'].value_counts().plot(kind='bar')

plt.xlabel("Category")

plt.ylabel("Count")

plt.title("Distribution of Categories")

plt.show()
 
```
